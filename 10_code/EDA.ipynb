```python
import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Exploratory Data Anlysis

```python
# Read the data
df = pd.read_csv("Combined_News_DJIA.csv")
```

```python
# Plot sentiment class distribution
barplot = sns.barplot(data = df, x = df["Label"].value_counts().index, y = df["Label"].value_counts().values)
plt.ylabel('Count')
plt.xlabel('Label')
plt.title('Sentiment Distribution')
plt.show()
fig = barplot.get_figure()
# fig.savefig('data_class_distribution.png')
```
